```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for the experiments
data = {
    "Model Type": [
        "XGBoost", "XGBoost", "XGBoost", "LSTM", "LSTM", "LSTM", "CNN-LSTM", "CNN-LSTM", "LSTM", "CNN-LSTM",
        "Transformer", "Transformer", "GRU", "GRU"
    ],
    "MAE": [
        0.0009, 0.0008, 0.0012, 0.0019, 0.0161, 0.0168, 0.0179, 0.0128, 0.0204, 0.019, 0.0127, 0.0147, 0.0128, 0.0115
    ],
    "RMSE": [
        1.0341, 1.0041, 1.2186, 3.0758, 0.0372, 0.0415, 0.0376, 0.0368, 0.0378, 0.0382, 0.0366, 0.0371, 0.0368, 0.0365
    ],
    "MSE": [
        1.0693, 1.0084, 1.485, 9.4609, 0.0013, 0.0017, 0.00142, 0.0014, 0.0014, 0.0014, 0.0013, 0.0014, 0.0014, 0.0014
    ],
    "R2": [
        0.997, 0.9973, 0.9959, 0.9974, 0.6874, 0.6108, 0.6794, 0.6943, 0.6709, 0.6699, 0.698, 0.6897, 0.694, 0.6985
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Group by Model Type and calculate the mean for each metric
grouped = df.groupby("Model Type").mean()

# Plot the metrics for each model type
plt.figure(figsize=(10, 6))
for metric in ["MAE", "RMSE", "MSE", "R2"]:
    plt.plot(grouped.index, grouped[metric], marker="o", label=metric)

plt.title("Scores for Each Model")
plt.xlabel("Model Type")
plt.ylabel("Scores")
plt.legend(title="Metrics")
plt.grid(True)
plt.tight_layout()
plt.show()
```
